```python
# ho bisogno di convertire la colonna dei casi in numerico 
struttura['total_cases'] = pd.to_numeric(struttura['total_cases'], errors='coerce')

# tolgo tutti i nulli in continent
struttura_continente = struttura.dropna(subset=['continent'])

# Raggruppa i dati per continente e trova il massimo dei casi totali per ogni continente
casi_per_continente = struttura_continente.groupby('continent')['total_cases'].max().reset_index()
# mi sono presa il massimo perchÃ¨ essendo continent un incrementale l'ultima data sarÃ  giÃ  una somma di tutti i casi 
```

â€¢Sempre riguardo i casi di COVID totali, si chiede di sviluppare una funzione che prenda in input il dataset e due nomi di continenti, e che ne confronti i seguenti relativi descrittori statistici: valori minimo e massimo, media, e percentuale rispetto al numero dei casi totali nel mondo (calcolati anche sulle locazioni senza indicazione di continente); 

```python

'''voglio una funzione che dia la possibilitÃ  di scegliere due contenti 
in base al continente scelto nella colonna continenti di struttura_continente mi
dia in output min, max, mean e percentuale dei casi '''

# due input 
# total cases
# descrizione statistica: min, max, mean, percent tot casi nel mondo) 

def confronta_continenti(struttura, continente1, continente2):
   #lavoro prima sul continente 1 
    # Filtro il dataset per includere solo i casi relativi al continente1 specificati
    casi_continente1= struttura[struttura['continent'] == continente1]['total_cases']

    # mi calcolo i dati statistici su continente1
    minimo_continente1 = casi_continente1.min()
    massimo_continente1 = casi_continente1.max()
    media_continente1 = casi_continente1.mean()
    # uso la somma dei nuovi casi (non uso total case perchÃ¨ Ã¨ incrementale) come variabile 
    # ho dovuvo impostare new cases come numerico 
    struttura['new_cases'] = pd.to_numeric(struttura['new_cases'], errors='coerce')
    casi_totali_mondo = struttura['new_cases'].sum()
    # Calcolo le percentuali rispetto ai casi totali nel mondo per continente1
    casi_continente1_1= struttura[struttura['continent'] == continente1]['new_cases']
    percentuale_continente1 = (casi_continente1_1.sum() / casi_totali_mondo) * 100
    
    #ripeto tutto su continente 2 
    # Filtro il dataset per includere solo i casi relativi al continente2 specificati
    casi_continente2 = struttura[struttura['continent'] == continente2]['total_cases']
    # mi calcolo i dati statistici su continente2
    minimo_continente2 = casi_continente2.min()
    massimo_continente2 = casi_continente2.max()
    media_continente2 = casi_continente2.mean()
    
    # Calcolo le percentuali rispetto ai casi totali nel mondo per continente2
    casi_continente2_2 = struttura[struttura['continent'] == continente2]['new_cases']
    percentuale_continente2 = (casi_continente2_2.sum() / casi_totali_mondo) * 100
    print("Il min di casi in", continente1, "Ã¨", minimo_continente1)
    print("Il min di casi in", continente2, "Ã¨", minimo_continente2)
    print("Il max di casi in", continente1, "Ã¨", massimo_continente1)
    print("Il max di casi in", continente2, "Ã¨", massimo_continente2)       
    print("La media dei casi in", continente1, "Ã¨", media_continente1)
    print("La media dei casi in", continente2, "Ã¨", media_continente2)  
    print("La percentuale dei casi in", continente1, "Ã¨", percentuale_continente1)
    print("La percentuale dei casi in", continente2, "Ã¨", percentuale_continente2) 
```

```python

confronta_continenti(struttura, 'Europe', 'South America')
```

â€¢Si chiede poi di effettuare lo stesso tipo di analisi â€“ anche in questo caso sviluppando una funzione ad hoc â€“ per il numero di vaccinazioni totali per ogni continente; 

```python
#basandomi sulla funzione di prima volevo che si ptoessero inserire piÃ¹ di due continenti selezionati 

def confronta_vaccinazioni_continenti2(struttura, *continenti):
    for continente in continenti:
        struttura['total_vaccinations'] = pd.to_numeric(struttura['total_vaccinations'], errors='coerce')
    
        vaccinazioni_continente = struttura[struttura['continent'] == continente]['total_vaccinations']

        min_vax_continente = vaccinazioni_continente.min()
        max_vax_continente = vaccinazioni_continente.max()
        mean_vax_continente = vaccinazioni_continente.mean()
        vax_totali_mondo = struttura['total_vaccinations'].sum()
        percentuale_vax_continente = (vaccinazioni_continente.sum() / vax_totali_mondo) * 100
    

        print("Il min dei vaccini in", continente, "Ã¨", min_vax_continente)
        print("Il max dei vaccini in", continente, "Ã¨", max_vax_continente)     
        print("La media dei vaccini in", continente, "Ã¨", mean_vax_continente)
        print("La percentuale dei vaccini in", continente, "Ã¨", percentuale_vax_continente)
```

```python
# prova per vedere se la funzione funzioni
confronta_vaccinazioni_continenti2 (struttura, 'Europe', 'South America', 'Asia')
```
